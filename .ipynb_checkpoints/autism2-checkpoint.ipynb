# INTRODUCTION:
Autism, also known as Autism Spectrum Disorder (ASD), is a developmental disorder that affects communication, behavior, and social interaction. The symptoms of autism can vary widely in terms of type and severity, which is why it is referred to as a "spectrum."

Key Characteristics of Autism:
1)Social Communication Challenges:
  -Difficulty with understanding and using verbal and non-verbal communication.
  -Trouble understanding social norms and cues.
  -Difficulty forming and maintaining relationships.
  -Repetitive Behaviors and Restricted Interests:

2)Engaging in repetitive movements or activities (e.g., hand-flapping, lining up toys).
  -Having highly focused interests or hobbies.
  -Insistence on sameness and routine.
  
3)Sensory Sensitivities:
  -Over- or under-reactivity to sensory stimuli such as lights, sounds, textures, or smells.

# FEATURES AND LABELS
 There is no ascertain way to tell whether a person has Autism or not because there are no such diagnostics methods available to diagnose this     disorder. But what if we use machine learning to predict whether a person suffers from Autism or not?
The features of our dataset are:
1) A1_score,A2_score....etc-These scores like represent to a series of questions/assessments related to autism screening. Each score indicate a binary response for specific traits or behaviour.
2) gender-female/male
3) relation-This represent the relation of the respondent with the individual being screened. Eg:parent,self etc
4) ethnicity
5) autism-This tells us whether an individual has a family member with ASD
6) used_app_before-whether the individual has used autism screening app before
7) result-This feature likely represents a numerical score or result from screening application,which could be composite score derived from responses to screening equations.

The label is class/ASD.Based on the above features,we will train our model. After training,our model should be able to predict for future unknown data whether a person has autism or not

# AUTISM vs AGE

```python
def convertAge(age):
    if age < 4:
        return 'Toddler'
    elif age < 12:
        return 'Kid'
    elif age < 18:
        return 'Teenager'
    elif age < 40:
        return 'Young'
    else:
        return 'Senior'
 
df['ageGroup'] = df['age'].apply(convertAge)
```

Here we can conclude that the Young and Toddler group of people have lower chances of having Autism

```python
sb.countplot(x=df['ageGroup'], hue=df['Class/ASD'])
plt.show()
```

```python
def add_feature(data):
   
  # Creating a column with all values zero
  data['sum_score'] = 0
  for col in data.loc[:,'A1_Score':'A10_Score'].columns:
     
    # Updating the 'sum_score' value with scores
    # from A1 to A10
    data['sum_score'] += data[col]
 
  # Creating a random data using the below three columns
  data['ind'] = data['austim'] + data['used_app_before'] + data['jaundice']
 
  return data
 
df = add_feature(df) 
```

```python
sb.countplot(x=df['sum_score'], hue=df['Class/ASD'])
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sb

def encode_labels(data):
    for col in data.columns:
        # Here we will check if datatype
        # is object then we will encode it
        if data[col].dtype == 'object':
            le = LabelEncoder()
            data[col] = le.fit_transform(data[col])
     
    return data

df = encode_labels(df)

# Making a heatmap to visualize the correlation matrix
plt.figure(figsize=(10,10))
sb.heatmap(df.corr() > 0.8, annot=True, cbar=False)
plt.show()
```

# MODEL TRAINING
Now we will train various models like:
1) LogisticRegression
2) RandomForestClassifier
3) KNeighborsClassifier
4) MLPClassifier
5) SVC
We"ll cpmpare their accuracy and check which model is fit for most close prediction for autism in an individual.

```python
removal = ['ID', 'age_desc', 'used_app_before', 'austim']
features = df.drop(removal + ['Class/ASD'], axis=1)
target = df['Class/ASD']
```

```python
from sklearn.model_selection import train_test_split

# Splitting the data into train and validation sets
X_train, X_val, Y_train, Y_val = train_test_split(features, target, test_size=0.2, random_state=10)

# Checking the shape of the split data
X_train.shape, X_val.shape, Y_train.shape, Y_val.shape
```

```python
from sklearn.preprocessing import MinMaxScaler

# Create MinMaxScaler object
scaler = MinMaxScaler()

# Fit scaler on training data and transform it
X_train_normalized = scaler.fit_transform(X_train)

# Transform validation data using the same scaler
X_val_normalized = scaler.transform(X_val)

# Check the shape of the normalized data
print("Shape of normalized training data:", X_train_normalized.shape)
print("Shape of normalized validation data:", X_val_normalized.shape)

# Reshape Y_train and Y_val to be 2-dimensional arrays
Y_train_reshaped = Y_train.values.reshape(-1, 1)
Y_val_reshaped = Y_val.values.reshape(-1, 1)

# Fit scaler on training target data and transform it
Y_train_normalized = scaler.fit_transform(Y_train_reshaped)

# Transform validation target data using the same scaler
Y_val_normalized = scaler.transform(Y_val_reshaped)

# Check the shape of the normalized target data
print("Shape of normalized training target data:", Y_train_normalized.shape)
print("Shape of normalized validation target data:", Y_val_normalized.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import roc_auc_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier

# Splitting the data into train and validation sets
X_train, X_val, Y_train, Y_val = train_test_split(features, target, test_size=0.2, random_state=10)

# Checking the shape of the split data
print("Shape of training data:", X_train.shape)
print("Shape of validation data:", X_val.shape)
print("Shape of training labels:", Y_train.shape)
print("Shape of validation labels:", Y_val.shape)

# Normalize the features
scaler = MinMaxScaler()
X_train_normalized = scaler.fit_transform(X_train)
X_val_normalized = scaler.transform(X_val)

# Initialize a list of models to compare
models = [
    LogisticRegression(),
    RandomForestClassifier(),
    KNeighborsClassifier(),
    MLPClassifier(),
    SVC(kernel='rbf')
]

# Train and evaluate each model
for model in models:
    model.fit(X_train_normalized, Y_train)
    train_predictions = model.predict(X_train_normalized)
    val_predictions = model.predict(X_val_normalized)
    
    print(f'{model.__class__.__name__}:')
    print('Training Accuracy:', roc_auc_score(Y_train, train_predictions))
    print('Validation Accuracy:', roc_auc_score(Y_val, val_predictions))
    print()
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

def plot_confusion_matrix(model, X, y):
    # Predict the labels
    y_pred = model.predict(X)
    # Generate the confusion matrix
    cm = confusion_matrix(y, y_pred)
    # Plot the confusion matrix using seaborn heatmap
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.xlabel('Predicted labels')
    plt.ylabel('True labels')
    plt.title('Confusion Matrix')
    plt.show()

# Usage
plot_confusion_matrix(models[0], X_val, Y_val)
```

# CONCLUSION
The machine learning model we have created is around 80% to 85% accurate. The disease for which there are no diagnostics methods machine learning models are able to predict whether the person has Autism or not.

The following conclusions can be drawn:
1) Logistic regression would be best suitable for predicting autism because they perform better on the validation data with less difference between the validation and training data. The second most suitable model would be SVC.

2) Random Forest Classifier would lead to overfitting. The model is too complex and has learned noise and details from the training data that do not generalize to new data.

3) KNeighborsClassifier: Shows a better balance between training and validation accuracies, indicating it might be more robust in generalizing to new data.
MLPClassifier: While it has higher training accuracy, the larger drop in validation accuracy suggests it might require more careful tuning or regularization to improve its performance on unseen data.
In conclusion, while both models show reasonably good accuracies, the KNeighborsClassifier appears to be performing slightly better in terms of generalization to unseen data based on the provided accuracy metrics. However, both models could benefit from further analysis and fine-tuning to optimize their performance and reduce overfitting.
